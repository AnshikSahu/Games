```python
def strategy(player_list,n):
    strat={}
    for k in player_list:
        player=players[k]
        prev="2"*n
        for i in range(0,len(player)):
            if(i%2==0):
                encodeing=encode(prev)
                if(encodeing in strat):
                    strat[encodeing].append(player[i])
                else:
                    strat[encodeing]=[player[i]]
            prev=prev[1:]+player[i]
    return strat
```

```python
def process_strat(strat):
    new_strat={}
    for key in strat:
        ones=strat[key].count("1")
        zeros=strat[key].count("0")
        if(ones>zeros):
            new_strat[key]=(1,ones/(ones+zeros))
        else:
            new_strat[key]=(0,zeros/(ones+zeros))
    return new_strat
```

```python
import numpy as np
from scipy.optimize import minimize_scalar

def is_concave(y_values):
    x_values=[i+1 for i in range(0, len(y_values))]
    # Check if function is concave
    n = len(x_values)
    if n < 3:
        return False  # Too few points to determine concavity

    x_values = np.array(x_values)
    y_values = np.array(y_values)

    # Compute slopes between consecutive points
    slopes = np.diff(y_values) / np.diff(x_values)

    # Check if slopes are non-increasing
    concaveity=0
    for i in range(1, len(slopes)):
        if slopes[i] < slopes[i - 1]:
            concaveity+=1
    if(concaveity>len(slopes)/2):
        return True
    else:
        return False

def find_minima(y_values):
    # Find minima of a concave function
    x_values=[i+1 for i in range(0, len(y_values))]

    # Define objective function for minimization
    def objective(x):
        return np.interp(x, x_values, y_values)

    # Find minimum using scalar minimization
    result = minimize_scalar(objective, bounds=(min(x_values), max(x_values)), method='bounded')
    return result.x
```
